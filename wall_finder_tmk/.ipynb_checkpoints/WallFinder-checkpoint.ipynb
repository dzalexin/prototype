```python
import torch
import torch.nn as nn
import torch.nn.functional as nnf
import torchvision.transforms as transforms
from PIL import Image
from skimage.morphology import medial_axis, skeletonize, thin
from skimage import feature
import matplotlib.pyplot as plt
import numpy as np

to_tensor = transforms.ToTensor()
to_image = transforms.ToPILImage()
```

# The Plan

The first is the typical wall pattern, the second is the same image rotated $\frac{\tau}{4}$ radians.

# The Process

What we're doing is "Good Old Fashioned AI" or "feature engineering", i.e., a human (me) figures out what features are important and extracts them to be placed into a function that can then find similar features in new images. It's almost as basic as it gets. We're using an ML library called `pytorch` to get a function called "convolution" which is an efficient way to compare the two kernels above with every point in the input image. The comparison happens with multiplication and we know that, if our kernel (<img src="kernel.png" width="16">) = $a$, then an exact multiplicative match would be $a^2$. So for every input point $x$ in the input image under convolution, we know if $\frac{x \cdot a}{a^2} = 1.0 \pm t | t \ll 1.0$ then we have a stong match with our kernel within tolerance $t$.

```python
def load_image(path):
    im = Image.open(path)
    return to_tensor(im).unsqueeze(0)

def invert(tensor):
    return (tensor - tensor.max()) * -1
```

```python
image = load_image(r"plan.png")#
image = invert(image)
kernel = load_image(r"kernel.png")#
kernel = invert(kernel)
kernelr = invert(load_image(r"kernel-r.png"))
kernel = torch.cat([kernel, kernelr], dim=(0))
kernel.shape
```

```python
# Do the convolution
conv = nnf.conv2d(image, kernel)
```

```python
# Filter for strong matches (~1.0)
b, c, h, w = conv.shape
tolerance = 0.6
result = conv.sum(dim=(1))
# Some of our values probably got really big, so normalize to be in the right range
result = result.div(result.max()).gt(tolerance).float()
```

```python
# Let's see the result
result_array = result.numpy()
print(result_array.min(), result_array.max())
fig, axes = plt.subplots(1, 2, figsize=(20, 20))
axes[0].imshow(result_array[0,...])
axes[1].imshow(Image.open("plan.png"), cmap="Greys")
```

```python
# The 8'4" dimension in the input image is ~310 pixels long, hence 8.3ft / 310px
sqf_per_px = (8.3 / 310) ** 2
feet_per_px = 8.3 / 310
```

```python
# What is the area that the walls cover?
wall_area = (conv.gt(0).sum() * sqf_per_px).item()
f"Wall area: {wall_area:.3f} sqft"
```

```python
# To find the lengths of all the walls, we take our strong matches and find the skeleton
skel, dist = medial_axis(result_array[0, :, :], return_distance=True)
# skel = skeletonize(result_array[0,...])
# skel = thin(result_array[0,...])
walls_length = dist.sum() * feet_per_px
print(f"Walls length: {walls_length:.3f} ft")
plt.imshow(skel)
```

```python
# Now let's find the perimeter length of each wall candidate we found through convolution
wall_edges = feature.canny(result_array[0,...], sigma=4.0)
plt.imshow(wall_edges)
```

```python
def gaussian_kernel(mu=0, sigma=1, width=30):
    x = np.linspace(-2 * sigma, 2 * sigma, width)
    y = np.linspace(-2 * sigma, 2 * sigma, width)
    
    xx, yy  = np.meshgrid(x, y)
    
    expx = np.e ** (-1 * (xx - mu) ** 2)/(2 * sigma ** 2)
    expy = np.e ** (-1 * (yy - mu) ** 2)/(2 * sigma ** 2)

    return ((1 / (sigma * np.sqrt(2 * np.pi))) * expx), ((1 / (sigma * np.sqrt(2 * np.pi))) * expy)
```

```python
gx, gy = gaussian_kernel(sigma=1, width=12)
```

```python
smoothed = nnf.conv2d(conv.unsqueeze(0), torch.from_numpy(gx * gy).reshape(1, 1, 12, 12).float()#)
```
